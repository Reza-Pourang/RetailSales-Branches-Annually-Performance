```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the datasets and Create an ExcelFile object
from google.colab import drive
drive.mount('/content/drive')
Data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Branch-Cleaned.csv')
```

```python
#Initial check
print(Data.head(5))
print('\n')
print(Data.dtypes)
print('\n')
print(Data.info())
```

```python
#Top States Based on Unique Customers - First Month

#Create a dictionary for renaming state labels
state_labels = {
    'زاهدان': 'Zahedan',
    'تبریز': 'Tabriz',
    'زنجان': 'Zanjan',
    'اصفهان': 'Isfahan',
    'ارومیه': 'Urmia',
    'اراک': 'Arak',
    'مرکزی': 'Markazi',
    'اردبیل': 'Ardebil',
    'کریمی': 'Karimi',
    'کرمان': 'Kerman',
    'کبیر پویان': 'KabirPouyan',
    'شوشتر': 'Shoushtar',
    'مشهد': 'Mashhad',
    'اهواز': 'Ahvaz',
    'گلستان': 'Golestan',
    'کاشان': 'Kashan',
    'گیلان': 'Gilan',
    'مازندران': 'Mazandaran',
    'کیش': 'Kish',}

#Rename state labels in the original DataFrame, creating a new column 'State_English'
Data['State_English'] = Data['State'].map(state_labels)

#Filter data for the first month
first_month_data = Data[Data["MonthNumeric"] == 1]

#Group by state and count unique customers using the new 'State_English' column
state_customer_counts = first_month_data.groupby("State_English")["CustomerCode"].nunique().reset_index()

#Sort by customer count in descending order and select the top 10 states
top_15_states = state_customer_counts.sort_values(by="CustomerCode", ascending=False).head(15)

#Plot the bar chart
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="State_English", y="CustomerCode", data=top_15_states, color="royalblue")  # Use 'State_English'
plt.xlabel("State")
plt.ylabel("Number of Unique Customers")
plt.title("Top States with Most Unique Customers in Month 1")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points')
plt.show()
```
